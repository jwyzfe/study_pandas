```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import FinanceDataReader as fdr

mpl.rcParams['font.family'] = 'NanumGothic'
```

```python
symbol_kospi = fdr.DataReader('KS11',start = '1990', end='2023')
symbol_shanghai = fdr.DataReader('SSEC',start = '1900', end='2023') 
symbol_japan = fdr.DataReader('N225',start = '1990', end='2023')
symbol_hangseng = fdr.DataReader('HSI',start = '1990', end='2023')
symbol_france = fdr.DataReader('FCHI',start = '1990', end='2023')

```

```python
series_kospi = (symbol_kospi['Close'] / symbol_kospi['Close'].iloc[0]) * 100
series_shanghai = (symbol_shanghai['Close'] / symbol_shanghai['Close'].iloc[0]) * 100
series_japan = (symbol_japan['Close'] / symbol_japan['Close'].iloc[0]) * 100
series_hangseng = (symbol_hangseng['Close'] / symbol_hangseng['Close'].iloc[0]) * 100
series_france = (symbol_france['Close'] / symbol_france['Close'].iloc[0]) * 100
```

```python

plt.plot(symbol_kospi.index, series_kospi, label='KOSPI', color='red')
plt.plot(symbol_shanghai.index, series_shanghai, label='SHANGHAI', color='blue' )
plt.plot(symbol_japan.index, series_japan, label='JAPAN', color='green')
plt.plot(symbol_hangseng.index, series_hangseng, label='HANGSENG', color='pink')
plt.plot(symbol_france.index, series_france, label='FRANCE', color='purple')
plt.legend()


plt.show()
```
