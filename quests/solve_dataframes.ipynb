```python
import pandas as pd
```

```python
data = {'종목 코드' : ['60310', '95570', '6840', '54620']
        , '종목명' : ['3S', 'AJ네트웍스', 'AK홀딩스', 'APS']
        , '시작일 기준가' : ['1890', '4600', '11200','5660']
        , '종료일 종가' : ['1941', '4745', '11070', '5700']
        , '대비' : ['51', '145', '-130', '40']
        , '등락률' : ['2.7','3.15','-1.16','0.71']
        , '거래량' : ['309268','104620','55481','208525']
        , '거래대금' : ['5595723413', '4806955935', '604085850', '1129272070']
}

df_components = pd.DataFrame(data)
df_components
```

```python
#열 총합 계산
df_components[['대비']].sum()
```

```python
df_components.set_index('종목 코드', inplace=True)
```

```python
df_selected = df_components[['종목명', '등락률']]
df_selected
```

```python
df_components['등락률'] = df_components['등락률'].astype(float)
avg_change_rate = df_components['등락률'].mean()
avg_change_rate
```

```python
positive_change = df_components[df_components['등락률'] > 0]
```

```python
df_components['거래대금'] = df_components['거래대금'].astype(float)
high_volume = df_components[df_components['거래대금'] >= 10000000000]


result = high_volume[['종목명', '거래대금']]
print(result)
```

```python
df_components['거래량'] = df_components['거래량'].astype(float)
as_sorted_by_volume = df_components.sort_values(by='거래량', ascending=False)

as_sorted_by_volume
```

```python
des_sorted_by_change_rate = df_components.sort_values(by='등락률', ascending=True)
des_sorted_by_change_rate
```

```python

df_components['시작일 기준가'] = df_components['시작일 기준가'].astype(float)
df_components['종료일 종가'] = df_components['종료일 종가'].astype(float)

df_components['수익률'] = (df_components['종료일 종가'] - df_components['시작일 기준가']) / df_components['시작일 기준가'] * 100
df_components
```
