```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from pykrx import stock
df_stock = stock.get_market_ohlcv_by_date('20000101', '20201231', '041510')
df_stock_close = df_stock[['종가']]
df_stock_close.head()
```

```python
df_stock.describe()
```

```python
df_stock_close['변화량'] = df_stock_close['종가'] - df_stock_close['종가'].shift(1)
df_stock_close
```

```python
condition_up = df_stock_close['변화량'] >= 0
condition_down = df_stock_close['변화량'] < 0
condition_down
```

```python
df_stock_close.loc[condition_up, '상승폭'] = df_stock_close['변화량']
df_stock_close.loc[condition_down, '하락폭'] = - df_stock_close['변화량']
df_stock_close
```

```python
df_stock_close = df_stock_close.fillna(0)
df_stock_close
```

```python
df_stock_close['AU'] = df_stock_close['상승폭'].rolling(window=14).mean()
df_stock_close['DU'] = df_stock_close['하락폭'].rolling(window=14).mean()
df_stock_close[13:20]
```

```python
df_stock_close['RSI'] = (df_stock_close['AU'] / (df_stock_close['AU'] + df_stock_close['DU'])) * 100
df_stock_close[12:].head()
```

```python
#매매 신호 설정
df_stock_close['매수'] = (df_stock_close['RSI'] < 30)  # RSI < 30
df_stock_close['매도'] = (df_stock_close['RSI'] > 70)  # RSI > 70
df_stock_close['보유'] = 1 - df_stock_close['매수'] - df_stock_close['매도']  # RSI 30 <= RSI <= 70
```

```python
# 수익률 계산 (전략과 보유 비교)
df_stock_close['수익률'] = df_stock_close['종가'].pct_change()  # 종가 수익률
df_stock_close['전략수익률'] = 0  # 전략 수익률 (초기화)
```

```python
# 전략 백테스팅: 매수/매도/보유에 따라 수익률 적용
for i in range(1, len(df_stock_close)):
    if df_stock_close['매수'].iloc[i] == 1:
        df_stock_close['전략수익률'].iloc[i] = df_stock_close['수익률'].iloc[i]  # 매수 시 수익률 반영
    elif df_stock_close['매도'].iloc[i] == 1:
        df_stock_close['전략수익률'].iloc[i] = -df_stock_close['수익률'].iloc[i]  # 매도 시 수익률 반영
    else:
        df_stock_close['전략수익률'].iloc[i] = df_stock_close['전략수익률'].iloc[i-1] + df_stock_close['수익률'].iloc[i]  # 보유 시 누적 수익률
```

```python
# 누적 수익률 계산
df_stock_close['누적수익률'] = (1 + df_stock_close['전략수익률']).cumprod()

# 단순 보유와 전략 비교
df_stock_close['누적수익률_단순보유'] = (1 + df_stock_close['수익률']).cumprod()

# 결과 출력
df_stock_close[['종가', 'RSI', '매수', '매도', '전략수익률', '누적수익률', '누적수익률_단순보유']].tail()
```
