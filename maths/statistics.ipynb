## 평균, 중앙값과 왜도와 관계

```python
# 데이터 생성 
data_symmetric = [1, 2, 3, 4, 5, 6, 7, 8, 9] # 대칭 분포 
data_positive_skew = [1, 2, 2, 3, 3, 3, 10, 10, 50] # 오른쪽 꼬리 
data_negative_skew = [-50, -10, -10, -3, -3, -3, -2, -2, -1] # 왼쪽 꼬리
```

```python
import numpy as np
from scipy.stats import skew, kurtosis #왜도 계산 함수
```

```python
mean = np.mean(data_positive_skew)
median = np.median(data_positive_skew)
skew = skew(data_positive_skew) #kurt > 0 -> 오른쪽 꼬리
mean, median, skew
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(data_positive_skew, bins = 10)  #단일 변수 분포 표시
plt.show()
```

```python
import seaborn as sns
```

```python
sns.kdeplot(data_positive_skew)
plt.title('kde plot')
plt.show()
```

```python

def calculate_statistics(data):
    
    mean_value = np.mean(data)
    median_value = np.median(data)
    skewness_value = skew(data)

    return mean_value, median_value, skewness_value


sns.kdeplot(data_positive_skew)
sns.kdeplot(data_symmetric)
sns.kdeplot(data_negative_skew)


```
