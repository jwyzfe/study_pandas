### visualization

### matplotlib

```python
import matplotlib.pyplot as plt
```

### 기본 구성

```python
x_data_first = [1,2,3,4]
y_data_first = [1,2,3,4]

x_data_second = [2,3,1,1]
y_data_second = [2,1,4,1]

x_data_third = [1,2,3,4]
y_data_third = [2,1,4,1]

x_data_fourth = [4,3,2,1]
y_data_fourth = [4,3,2,1]
```

### 정렬
- add_subplot, subplots()

```python
# fig, axes = plt.subplots(2,2, figsize=(14,6))
fig, axes = plt.subplots(1,2)

axes[0].plot(x_data_first, y_data_first)
axes[1].plot(x_data_third, y_data_third)

plt.show()
```

### 구성요소 추가


```python
# fig, axes = plt.subplots(2,2, figsize=(14,6))
fig, axes = plt.subplots(1,2)

fig.suptitle('Super Title')

axes[0].plot(x_data_first, y_data_first, label = 'legend first')
axes[0].plot(x_data_third, y_data_third, label = 'legend second')
axes[0].set_title('axes first')
axes[0].set_xlabel('first x_data')
axes[0].legend

axes[1].plot(x_data_third, y_data_third)
axes[1].set_title('axes second')
axes[1].set_ylabel('second y_data')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1,4, figsize = (14,6))

axes[0].plot(x_data_first, y_data_first)
axes[1].plot(x_data_third, y_data_third)
axes[2].plot(x_data_fourth, y_data_fourth)
axes[3].plot(x_data_second, y_data_second)

plt.show()
```

### figure size 조정

```python

fig, axes = plt.subplots(1,2, figsize=(10,6), dpi = 100)

fig.suptitle('Super Title')

axes[0].plot(x_data_first, y_data_first, label = 'legend first')
axes[0].plot(x_data_third, y_data_third, label = 'legend second')
axes[0].set_title('axes first')
axes[0].set_xlabel('first x_data')
axes[0].legend

axes[1].plot(x_data_third, y_data_third)
axes[1].set_title('axes second')
axes[1].set_ylabel('second y_data')

plt.tight_layout()
plt.show()
```
