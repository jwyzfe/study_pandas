```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import FinanceDataReader as fdr

# 한글 폰트 설정 (NanumGothic) 
mpl.rcParams['font.family'] = 'NanumGothic'
```

```python
# API 사용
import FinanceDataReader as fdr
symbol_kospi = fdr.DataReader('KS11', '2023')
symbol_kospi
```

```python
plt.figure(figsize=(10,6), dpi = 100)

fig, axes = plt.subplots(1,4, figsize = (14,6))


# Open vs High
axes[0].plot(symbol_kospi.index, symbol_kospi['Open'], label='Open')
axes[0].plot(symbol_kospi.index, symbol_kospi['High'], label='High')
axes[0].set_title('Open vs High')
axes[0].legend()

# Open vs Low
axes[1].plot(symbol_kospi.index, symbol_kospi['Open'], label='Open')
axes[1].plot(symbol_kospi.index, symbol_kospi['Low'], label='Low')
axes[1].set_title('Open vs Low')
axes[1].legend()

# Open vs Close
axes[2].plot(symbol_kospi.index, symbol_kospi['Open'], label='Open')
axes[2].plot(symbol_kospi.index, symbol_kospi['Close'], label='Close')
axes[2].set_title('Open vs Close')
axes[2].legend()

# Open vs Volume
axes[3].plot(symbol_kospi.index, symbol_kospi['Open'], label='Open')
axes[3].plot(symbol_kospi.index, symbol_kospi['Volume'], label='Volume')
axes[3].set_title('Open vs Volume')
axes[3].legend()


plt.suptitle('2023년 이후 코스피 지수')
fig.supxlabel('날짜')
plt.ylabel('지수')
plt.legend()
plt.show()
```
