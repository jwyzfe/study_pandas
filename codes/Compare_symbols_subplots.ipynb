```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import FinanceDataReader as fdr

# 한글 폰트 설정 (NanumGothic) 
mpl.rcParams['font.family'] = 'NanumGothic'
```

```python
# API 사용
import FinanceDataReader as fdr
kospi = fdr.DataReader('KS11', '2023-01-01')  # 코스피 지수
fertilizer = fdr.DataReader('025860', '2023-01-01')  # 남해화학 (비료 관련)
```

```python

# 서브플롯 생성
fig, axes = plt.subplots(4, 1, figsize=(12, 16))  # 4행 1열 서브플롯

# 날짜 간격 조정
date_indices = kospi.index[::10]  # 10일 간격으로 설정
date_labels = date_indices.strftime('%Y-%m-%d')  # 날짜 형식

# 그래프 1: Open 가격 비교
axes[0].plot(kospi.index, kospi['Open'], label='KOSPI Open')
axes[0].plot(fertilizer.index, fertilizer['Open'], label='남해화학 Open')
axes[0].set_title('Open 가격 비교', fontsize=14)
axes[0].legend()

# 그래프 2: High 가격 비교
axes[1].plot(kospi.index, kospi['High'], label='KOSPI High', color='blue', alpha=0.7)
axes[1].plot(fertilizer.index, fertilizer['High'], label='남해화학 High', color='green', alpha=0.7)
axes[1].set_title('High 가격 비교', fontsize=14)
axes[1].legend()

# 그래프 3: Low 가격 비교
axes[2].plot(kospi.index, kospi['Low'], label='KOSPI Low', color='blue', alpha=0.7)
axes[2].plot(fertilizer.index, fertilizer['Low'], label='남해화학 Low', color='green', alpha=0.7)
axes[2].set_title('Low 가격 비교', fontsize=14)
axes[2].legend()

# 그래프 4: Close 가격 비교
axes[3].plot(kospi.index, kospi['Close'], label='KOSPI Close', color='blue', alpha=0.7)
axes[3].plot(fertilizer.index, fertilizer['Close'], label='남해화학 Close', color='green', alpha=0.7)
axes[3].set_title('Close 가격 비교', fontsize=14)

axes[3].legend()

# 공통 제목 및 레이아웃 설정
fig.suptitle('코스피와 비료 관련 종목(남해화학) 비교', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.96])  # 공통 제목 영역 확보
plt.show()
```
