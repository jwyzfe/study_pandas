```python
import matplotlib.pyplot as plt
```

```python
x_data_first = [1,2,3,4]
y_data_first = [1,2,3,4]

x_data_second = [2,3,1,1]
y_data_second = [2,1,4,1]

x_data_third = [1,2,3,4]
y_data_third = [2,1,4,1]

x_data_fourth = [4,3,2,1]
y_data_fourth = [4,3,2,1]
```

### 단일변수 표현

```python
plt.plot(x_data_first)
plt.show()
```

### 다변수 표현

### 설명 요소 구성

```python
plt.plot(x_data_first,y_data_first, label='x_data_legend')
plt.plot(y_data_third, x_data_third, label='y_data_legend')

plt.title('visualization with Title')
plt.xlabel('x_data')
plt.ylabel('y_data')
plt.legend()

plt.show()
#tick,label,title
```

### 한글 처리

```python
import matplotlib as mpl 
# 한글 폰트 설정 (NanumGothic) 
mpl.rcParams['font.family'] = 'NanumGothic'
```

```python
plt.plot(x_data_first,y_data_first, label='x_data_범례')
plt.plot(y_data_third, x_data_third, label='y_data_범례')

plt.title('visualization with 타이틀')
plt.xlabel('x_데이터')
plt.ylabel('y_데이터')
plt.legend()

plt.savefig('visualizations_plots.png')
plt.show()
```

### figure 넓이, 높이 조정
